```python

import re

def process_text(text):
    # Заменяем ? на " ? " и ! на " ! "
    text = re.sub(r'([?!,\.])', r' \1 ', text)
    # Заменяем все остальные знаки препинания на пробелы
    text = re.sub(r"\\n", " ", text)
    text = re.sub(r'[^\w\s?!]', ' ', text)
    # Убираем лишние пробелы
    text = re.sub(r'\s+', ' ', text).strip()
    return text.lower()
```

Сравнение моделей

```python
from sklearn.neighbors import NearestNeighbors

# Создаем структуру для хранения всех моделей
models = {}

# Модель 1: Только текст (уже есть у вас)
models['text_only'] = {
    'name': 'Только текст',
    'vectorizer': vectorizer,
    'tfidf_matrix': index_tfidfs,
    'id_to_url': id_to_url,
    'data_column': 'text_processed',
    'max_features': 10000
}

# Модель 2: Только заголовок (уже есть у вас)
models['title_only'] = {
    'name': 'Только заголовок',
    'vectorizer': vectorizer_title,
    'tfidf_matrix': index_tfidfs_title,
    'id_to_url': id_to_url_title,
    'data_column': 'title_processed',
    'max_features': 10000
}

# Модель 3: Заголовок + текст (уже есть у вас)
models['title_text'] = {
    'name': 'Заголовок + текст',
    'vectorizer': vectorizer_title_text,
    'tfidf_matrix': index_tfidfs_title_text,
    'id_to_url': id_to_url_title_text,
    'data_column': 'title_text_processed',
    'max_features': 10000
}

# Создаем KNN индексы для каждой модели
for model_name, model in models.items():
    index_knn = NearestNeighbors(n_neighbors=5, metric='cosine')
    index_knn.fit(model['tfidf_matrix'])
    models[model_name]['index_knn'] = index_knn
```

Метрики

```python
def evaluate_model(model, questions_df, k=5):
    """
    Оценивает модель на всех вопросах
    """
    vectorizer = model['vectorizer']
    index_knn = model['index_knn']
    id_to_url = model['id_to_url']

    recalls_at_1 = []
    recalls_at_5 = []
    dcg_at_5_list = []

    for _, item in questions_df.iterrows():
        # Обработка вопроса
        query_processed = process_text(item["question"])
        query_embed = vectorizer.transform([query_processed])

        # Поиск документов
        distances, neighbour_ids = index_knn.kneighbors(query_embed[0])

        # Получаем URL найденных документов
        correct_url = item["correct_url"]
        neighbour_urls = [id_to_url[idx] for idx in neighbour_ids[0]]

        # Recall@1
        recall_at_1 = 1.0 if neighbour_urls[0] == correct_url else 0.0
        recalls_at_1.append(recall_at_1)

        # Recall@5
        recall_at_5 = 1.0 if correct_url in neighbour_urls[:k] else 0.0
        recalls_at_5.append(recall_at_5)

        # DCG@5
        dcg = 0.0
        for i, url in enumerate(neighbour_urls[:k]):
            relevance = 1.0 if url == correct_url else 0.0
            if i == 0:
                dcg += relevance
            else:
                dcg += relevance / np.log2(i + 1)
        dcg_at_5_list.append(dcg)

    return {
        'recall@1': np.mean(recalls_at_1),
        'recall@5': np.mean(recalls_at_5),
        'dcg@5': np.mean(dcg_at_5_list),
        'recall@1_std': np.std(recalls_at_1),
        'recall@5_std': np.std(recalls_at_5)
    }
```

Оценка

```python
# Оцениваем каждую модель
import numpy as np
results = {}
for model_name, model in models.items():
    print(f"Оценка модели: {model['name']}...")
    metrics = evaluate_model(model, questions_and_answers)
    results[model_name] = metrics

    print(f"  Recall@1: {metrics['recall@1']:.4f}")
    print(f"  Recall@5: {metrics['recall@5']:.4f}")
    print(f"  DCG@5:    {metrics['dcg@5']:.4f}")
    print()
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "Qwen/Qwen3-0.6B"

#model_name = "Qwen/Qwen3-4B"

# load the tokenizer and the model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
```

```python
def create_llm_answer(message, model, tokenizer, enable_thinking=False):
    messages = [
        {"role": "user", "content": message}
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=enable_thinking
    )

    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

    # conduct text completion
    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=4000
    )
    output_ids = generated_ids[0][len(model_inputs.input_ids[0]):].tolist()

    # parsing thinking content
    try:
        # rindex finding 151668 (</think>)
        index = len(output_ids) - output_ids[::-1].index(151668)
    except ValueError:
        index = 0

    thinking_content = tokenizer.decode(output_ids[:index], skip_special_tokens=True).strip("\n")
    content = tokenizer.decode(output_ids[index:], skip_special_tokens=True).strip("\n")

    return content
```

```python
from tqdm.notebook import tqdm
import pandas as pd
import numpy as np

# Функция для вычисления recall@k
def compute_recall_at_k(retrieved_urls, correct_urls, k=1):
    top_k = set(retrieved_urls[:k])
    return len(top_k & correct_urls)

# Выбираем лучшую модель — по результатам это 'title_text'
best_model = models['title_text']
vectorizer = best_schema_vectorizer = best_model['vectorizer']
index_knn = best_model['index_knn']
id_to_url = best_model['id_to_url']

llm_answers = []
MAX_QUESTIONS_TO_PROCESS = 50
start_from = 450

# Итерируемся по вопросам с прогресс-баром
for idx, item in tqdm(
    questions_and_answers.iloc[start_from:start_from + MAX_QUESTIONS_TO_PROCESS].iterrows(),
    total=MAX_QUESTIONS_TO_PROCESS
):
    # 1. Получаем релевантные документы
    query_processed = process_text(item["question"])
    query_embed = vectorizer.transform([query_processed])
    distances, neighbour_ids = index_knn.kneighbors(query_embed[0])

    neighbour_urls = [id_to_url[n_id] for n_id in neighbour_ids[0]]
    correct_urls = {item["correct_url"]}

    recall_at_1 = compute_recall_at_k(neighbour_urls, correct_urls, k=1)
    recall_at_5 = compute_recall_at_k(neighbour_urls, correct_urls, k=5)

    # 2. Генерация ответов

    # Без RAG
    llm_answer = create_llm_answer(item["question"], model, tokenizer)

    # RAG: 1 документ
    top1_doc_text = data.loc[neighbour_urls[0], "text"]
    message_rag1 = (
        f"Ответь на вопрос, используя только информацию ниже.\n\n"
        f"Контекст:\n{top1_doc_text}\n\n"
        f"Вопрос: {item['question']}"
    )
    llm_answer_rag1 = create_llm_answer(message_rag1, model, tokenizer)

    # RAG: 5 документов
    rag_docs = []
    for i, url in enumerate(neighbour_urls[:5], 1):
        doc_text = data.loc[url, "text"]
        rag_docs.append(f"Документ {i}:\n{doc_text}")
    message_rag5 = (
        f"Ответь на вопрос, используя информацию из следующих документов.\n\n"
        + "\n\n".join(rag_docs) + "\n\n"
        + f"Вопрос: {item['question']}"
    )
    llm_answer_rag5 = create_llm_answer(message_rag5, model, tokenizer)

    # 3. Сохраняем результат
    llm_answers.append({
        "question": item["question"],
        "correct_answer": item["correct_answer"],
        "llm_answer": llm_answer,
        "llm_answer_rag1": llm_answer_rag1,
        "llm_answer_rag5": llm_answer_rag5,
        "is_top1_doc_rel": bool(recall_at_1 > 0),
        "has_top5_doc_rel": bool(recall_at_5 > 0),
        "correct_url": item["correct_url"],
        "retrieved_urls_top5": "; ".join(neighbour_urls[:5])
    })

# 4. Сохраняем в Excel для ручной разметки
df_to_annotate = pd.DataFrame(llm_answers)
df_to_annotate.to_excel("llm_rag_annotation_50_samples.xlsx", index=False)

print("Файл сохранён: llm_rag_annotation_50_samples.xlsx")
print("Теперь откройте его в Excel и добавьте колонки:")
print(" - 'correct_no_rag' (1/0)")
print(" - 'correct_rag1' (1/0)")
print(" - 'correct_rag5' (1/0)")
print(" - 'rag_improved' (да/нет/не хуже)")
```

```python
import re

def get_judge_score_final(question: str, reference_answer: str, model_answer: str, model, tokenizer) -> str:
    """
    Надёжная оценка: извлекает 0/1 ТОЛЬКО из начала сгенерированной модели частью (после промпта).
    Защищена от ложных цифр вроде '1984' → '1'.
    """
    prompt = f"""Ты — строгий эксперт-оценщик. Твоя задача — сравнить два ответа и вывести ТОЛЬКО ОДНУ ЦИФРУ без каких-либо объяснений.


Пример:
ВОПРОС: Сколько золотых медалей у Майкла Фелпса?
ЭТАЛОННЫЙ ОТВЕТ: 23
ОТВЕТ МОДЕЛИ: У Майкла Фелпса 23 золотые медали.
Цифра: 1

Пример:
ВОПРОС: В каком году родился Иан Торп?
ЭТАЛОННЫЙ ОТВЕТ: 1982
ОТВЕТ МОДЕЛИ: Он родился в конце 1983.
Цифра: 0

Пример:
ВОПРОС: Кто стала чемпионкой олимпиады 2012 в Лондоне по плаванию вольным стилемм 50 м среди женщин?
ЭТАЛОННЫЙ ОТВЕТ: Раноми Кромовидьойо
ОТВЕТ МОДЕЛИ: Это был Майкл Фелпс.
Цифра: 0


ВОПРОС: {question}

ЭТАЛОННЫЙ ОТВЕТ: {reference_answer}

ОТВЕТ МОДЕЛИ: {model_answer}

Если ответ модели СУЩЕСТВЕННО соответствует эталонному ответу по смыслу и содержит правильную информацию, выведи: 1
Если ответ модели НЕ соответствует эталону, содержит ошибки или неточности, выведи: 0

Цифра:"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=5,
        do_sample=False,
        temperature=0.0,
        pad_token_id=tokenizer.eos_token_id
    )

    generated_tokens = outputs[0][inputs.input_ids.shape[1]:]
    generated_text = tokenizer.decode(generated_tokens, skip_special_tokens=True).strip()


    for char in generated_text:
        if char == '0':
            return '0'
        elif char == '1':
            return '1'
        elif char.isspace():
            continue
        else:
            break

    match = re.search(r'^\s*([01])', generated_text)
    if match:
        return match.group(1)
    return 'ERROR'
```

```python
import pandas as pd
import re

# Загрузка размеченных данных
df = pd.read_excel("llm_rag_annotation_50_samples.xlsx")

def safe_judge(*args, **kwargs):
    """Обёртка для обработки ошибок и некорректных выходов"""
    result = get_judge_score_final(*args, **kwargs)
    if result in ('0', '1'):
        return int(result)
    else:
        print(f" Судья вернул некорректное значение: '{result}' → заменено на 0")
        return 0  # или np.nan, если предпочитаете
```

```python
# Применяем LLM-as-a-judge ко всем трём типам ответов
print("Оценка ответов без RAG...")
df["llm_judge_no_rag"] = df.apply(
    lambda row: safe_judge(row["question"], row["correct_answer"], row["llm_answer"], model, tokenizer),
    axis=1
)

print("Оценка ответов с RAG-1...")
df["llm_judge_rag1"] = df.apply(
    lambda row: safe_judge(row["question"], row["correct_answer"], row["llm_answer_rag1"], model, tokenizer),
    axis=1
)

print("Оценка ответов с RAG-5...")
df["llm_judge_rag5"] = df.apply(
    lambda row: safe_judge(row["question"], row["correct_answer"], row["llm_answer_rag5"], model, tokenizer),
    axis=1
)

# Сохраняем результат
df.to_excel("llm_rag_with_judge_scores.xlsx", index=False)
print("Файл сохранён: llm_rag_with_judge_scores.xlsx")
```

```python
from sklearn.metrics import accuracy_score, cohen_kappa_score

# Без RAG
acc_no_rag = accuracy_score(df["correct_no_rag"], df["llm_judge_no_rag"])
kappa_no_rag = cohen_kappa_score(df["correct_no_rag"], df["llm_judge_no_rag"])

# RAG-1
acc_rag1 = accuracy_score(df["correct_rag1"], df["llm_judge_rag1"])
kappa_rag1 = cohen_kappa_score(df["correct_rag1"], df["llm_judge_rag1"])

# RAG-5
acc_rag5 = accuracy_score(df["correct_rag5"], df["llm_judge_rag5"])
kappa_rag5 = cohen_kappa_score(df["correct_rag5"], df["llm_judge_rag5"])

print("Совпадение LLM-as-a-judge с человеком:")
print(f"Без RAG:     Accuracy = {acc_no_rag:.2%}, Kappa = {kappa_no_rag:.2f}")
print(f"RAG-1:       Accuracy = {acc_rag1:.2%}, Kappa = {kappa_rag1:.2f}")
print(f"RAG-5:       Accuracy = {acc_rag5:.2%}, Kappa = {kappa_rag5:.2f}")

# Среднее
avg_acc = np.mean([acc_no_rag, acc_rag1, acc_rag5])
avg_kappa = np.mean([kappa_no_rag, kappa_rag1, kappa_rag5])
print(f"\nСреднее:     Accuracy = {avg_acc:.2%}, Kappa = {avg_kappa:.2f}")
```

```python
import random
import pandas as pd
from tqdm.notebook import tqdm

subset_questions = questions_and_answers.iloc[450:500].reset_index(drop=True)

all_urls = set(data.index)
results=[]
```

```python
for _, item in tqdm(subset_questions.iterrows(), total=len(subset_questions)):
    question = item["question"]
    correct_answer = item["correct_answer"]
    correct_url = item["correct_url"]

    # Исключаем правильный URL, выбираем 4 случайных нерелевантных
    irrelevant_pool = list(all_urls - {correct_url})
    if len(irrelevant_pool) < 4:
        continue
    random_urls = random.sample(irrelevant_pool, 4)

    correct_text = data.loc[correct_url, "text"]
    random_texts = [data.loc[url, "text"] for url in random_urls]

    # --- Вариант 1: релевантный на 1 месте ---
    docs1 = [correct_text] + random_texts[:4]
    context1 = "\n\n".join(f"Документ {i+1}:\n{text}" for i, text in enumerate(docs1))
    prompt1 = f"Ответь на вопрос, используя информацию из следующих документов.\n\n{context1}\n\nВопрос: {question}"
    ans1 = create_llm_answer(prompt1, model, tokenizer)

    # --- Вариант 2: релевантный на 3 месте ---
    docs3 = [random_texts[0], random_texts[1], correct_text, random_texts[2], random_texts[3]]
    context3 = "\n\n".join(f"Документ {i+1}:\n{text}" for i, text in enumerate(docs3))
    prompt3 = f"Ответь на вопрос, используя информацию из следующих документов.\n\n{context3}\n\nВопрос: {question}"
    ans3 = create_llm_answer(prompt3, model, tokenizer)

    # --- Вариант 3: релевантный на 5 месте ---
    docs5 = random_texts[:4] + [correct_text]
    context5 = "\n\n".join(f"Документ {i+1}:\n{text}" for i, text in enumerate(docs5))
    prompt5 = f"Ответь на вопрос, используя информацию из следующих документов.\n\n{context5}\n\nВопрос: {question}"
    ans5 = create_llm_answer(prompt5, model, tokenizer)

    # --- Оценка через LLM-as-a-judge ---
    score1 = safe_judge(question, correct_answer, ans1, model, tokenizer)
    score3 = safe_judge(question, correct_answer, ans3, model, tokenizer)
    score5 = safe_judge(question, correct_answer, ans5, model, tokenizer)

    results.append({
        "question": question,
        "correct_answer": correct_answer,
        "correct_url": correct_url,
        "llm_answer_pos1": ans1,
        "llm_answer_pos3": ans3,
        "llm_answer_pos5": ans5,
        "judge_pos1": score1,
        "judge_pos3": score3,
        "judge_pos5": score5
    })
```

```python
import wikipediaapi
import time
import json

# Настройка
USER_AGENT = "BasketballParser/1.0"
wiki_wiki = wikipediaapi.Wikipedia(
    language='ru',
    extract_format=wikipediaapi.ExtractFormat.WIKI,
    user_agent=USER_AGENT
)

def get_pages_from_categories(category_list, max_total_pages=110):
    """
    Получает страницы из нескольких категорий.

    Args:
        category_list: список названий категорий (без префикса 'Категория:')
        max_total_pages: максимальное общее количество страниц

    Returns:
        Список названий страниц
    """
    all_page_titles = []
    seen_titles = set()  # Для избежания дубликатов

    for category_name in category_list:
        print(f"Обрабатываю категорию: {category_name}")
        category = wiki_wiki.page(f"Категория:{category_name}")

        if not category.exists():
            print(f"  Категория '{category_name}' не найдена, пропускаю.")
            continue

        # Получаем страницы из этой категории
        for page in category.categorymembers.values():
            if len(all_page_titles) >= max_total_pages:
                break

            # Берем только основные статьи (пространство имен 0)
            if page.ns == wikipediaapi.Namespace.MAIN:
                if page.title not in seen_titles:
                    seen_titles.add(page.title)
                    all_page_titles.append(page.title)

        time.sleep(0.3)

    return all_page_titles

def main():
    # Список категорий для сбора статей о баскетболе
    basketball_categories = [
        "Баскетбол",
        "Баскетболисты_знаменитые",
        "Баскетбольные_клубы",
        "Баскетбольные_турниры_знаменитые",
        "НБА",
        "Баскетбольные_тренеры_знаменитые",
        "Чемпионаты_мира_по_баскетболу_среди_взрослых",
        "Олимпийские_чемпионы_по_баскетболу",
        "Чемпионат_Европы_по_баскетболу_среди_взрослых",
    ]

    print("Начинаю сбор статей о баскетболе...")
    print(f"Категории: {basketball_categories}")

    # Получаем список всех статей
    article_titles = get_pages_from_categories(
        basketball_categories,
        max_total_pages=110
    )

    print(f"\nНайдено {len(article_titles)} уникальных статей.")
    print(f"Примеры первых 10 статей: {article_titles[:10]}")

    documents = []
    error_log = []

    for i, title in enumerate(article_titles):
        if len(documents) >= 111:
            print(f"Достигнут лимит в 110 документов.")
            break

        print(f"[{i+1}/{len(article_titles)}] Парсинг: {title}")

        try:
            page = wiki_wiki.page(title)

            if not page.exists():
                error_log.append(f"Страница не существует: {title}")
                continue

            text = page.text.strip()

            if len(text) < 300:  # Минимальная длина
                print(f"  Статья слишком короткая ({len(text)} символов), пропускаю.")
                error_log.append(f"Слишком короткая статья: {title} ({len(text)} символов)")
                continue

            basketball_keywords = ['баскетбол', 'баскетболист', 'НБА', 'чемпионат', 'матч', 'очко', 'мяч']
            text_lower = text.lower()

            # Если статья не содержит ключевых слов баскетбола, проверяем ее содержание
            if not any(keyword in text_lower for keyword in basketball_keywords[:3]):
                # Проверяем начало статьи (обычно там определение)
                first_paragraph = text_lower[:500]
                if 'баскетбол' not in first_paragraph:
                    print(f"  Статья не связана с баскетболом, пропускаю.")
                    error_log.append(f"Не релевантная статья: {title}")
                    continue

            # Создаем документ
            doc = {
                "id": f"basketball_{len(documents) + 1:03d}",
                "title": title,
                "url": page.fullurl,
                "text": text
            }

            documents.append(doc)
            print(f" Добавлено (символов: {len(text)})")

        except Exception as e:
            error_msg = f"Ошибка при обработке {title}: {str(e)}"
            print(f" {error_msg}")
            error_log.append(error_msg)

        time.sleep(0.5)

    output_file = "wikipedia_basketball_docs_110.json"
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(documents, f, ensure_ascii=False, indent=2)

    if error_log:
        with open("parsing_errors.log", 'w', encoding='utf-8') as f:
            f.write("\n".join(error_log))

    print(f"\n=== РЕЗУЛЬТАТЫ ===")
    print(f"Успешно сохранено документов: {len(documents)}")
    print(f"Ошибок/пропущено статей: {len(error_log)}")
    print(f"Файл с документами: {output_file}")

if __name__ == "__main__":
    main()
```

```python
# Функция генерации QA
def generate_qa_with_deepseek(title: str, text: str, max_retries=3) -> Tuple[str, str]:

    # Ограничиваем длину текста для экономии токенов
    truncated_text = text[:2000].strip()
    if len(text) > 2000:
        truncated_text += " [...]"

    # Более чёткий промпт с примерами
    prompt = f"""Ты помощник, который создаёт вопросы и ответы на основе текста.

Заголовок статьи: {title}
Текст статьи: {truncated_text}

Создай ОДИН конкретный вопрос и ОДИН точный ответ на основе текста выше.
Вопрос должен быть на русском языке и касаться баскетбола.
Ответ должен быть кратким (1-2 предложения) и точным.

Формат вывода (обязательно придерживайся этого формата):
Вопрос: [здесь твой вопрос]
Ответ: [здесь твой ответ]

Пример:
Вопрос: В каком году была основана Национальная баскетбольная ассоциация?
Ответ: НБА была основана 6 июня 1946 года в Нью-Йорке.

Теперь создай вопрос и ответ для статьи "{title}":"""

    for attempt in range(max_retries):
        try:
            data = {
                "model": "deepseek-chat",
                "messages": [
                    {
                        "role": "system",
                        "content": "Ты эксперт по баскетболу. Создавай точные вопросы и ответы на основе предоставленных текстов."
                    },
                    {
                        "role": "user",
                        "content": prompt
                    }
                ],
                "max_tokens": 300,
                "temperature": 0.3,
                "top_p": 0.9,
                "frequency_penalty": 0.5
            }


            response = requests.post(
                DEEPSEEK_API_URL,
                headers=headers,
                json=data,
                timeout=45
            )

            # Проверяем статус
            if response.status_code == 200:
                result = response.json()
                output = result["choices"][0]["message"]["content"].strip()

                # Парсинг результата
                question, answer = "", ""
                lines = output.split("\n")

                for line in lines:
                    line = line.strip()
                    if line.startswith("Вопрос:") or line.startswith("Вопрос :"):
                        question = line.split(":", 1)[1].strip()
                    elif line.startswith("Ответ:") or line.startswith("Ответ :"):
                        answer = line.split(":", 1)[1].strip()

                if question and answer:
                    return question, answer
                else:
                    print(f"Не удалось распарсить ответ (попытка {attempt+1})")
                    print(f"     Ответ API: {output[:100]}...")

            elif response.status_code == 429:
                wait_time = min(2 ** (attempt + 1), 30)
                print(f" Rate limit. Жду {wait_time} секунд...")
                time.sleep(wait_time)
                continue

            else:
                print(f" Ошибка API (статус {response.status_code}): {response.text[:200]}")

        except requests.exceptions.Timeout:
            print(f"Таймаут (попытка {attempt+1})")
        except requests.exceptions.ConnectionError:
            print(f"Ошибка соединения (попытка {attempt+1})")
        except Exception as e:
            print(f"Неожиданная ошибка (попытка {attempt+1}): {e}")

        if attempt < max_retries - 1:
            time.sleep(2)

    # Если все попытки не удались
    return f"Что такое {title} в баскетболе?", "Не удалось сгенерировать ответ."
```

```python
# Функция оценки
def evaluate_model_on_questions(vectorizer, knn_index, id_to_url, questions_df, k=5):
    recalls_at_1 = []
    recalls_at_5 = []
    dcg_at_5_list = []

    for _, item in questions_df.iterrows():
        # Обработка вопроса
        query_processed = process_text(item["question"])
        query_vec = vectorizer.transform([query_processed])

        # Поиск ближайших соседей
        distances, neighbour_ids = knn_index.kneighbors(query_vec)

        # Получаем URL
        correct_url = item["correct_url"]
        retrieved_urls = [id_to_url[idx] for idx in neighbour_ids[0]]

        # Recall@1
        recall_at_1 = 1.0 if retrieved_urls[0] == correct_url else 0.0
        recalls_at_1.append(recall_at_1)

        # Recall@5
        recall_at_5 = 1.0 if correct_url in retrieved_urls[:k] else 0.0
        recalls_at_5.append(recall_at_5)

        # DCG@5
        dcg = 0.0
        for rank, url in enumerate(retrieved_urls[:k]):
            relevance = 1.0 if url == correct_url else 0.0
            if rank == 0:
                dcg += relevance
            else:
                dcg += relevance / np.log2(rank + 1)
        dcg_at_5_list.append(dcg)

    return {
        'recall@1': np.mean(recalls_at_1),
        'recall@5': np.mean(recalls_at_5),
        'dcg@5': np.mean(dcg_at_5_list)
    }
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F
import numpy as np
from tqdm.notebook import tqdm

# Загрузка модели и токенизатора
model_name = "intfloat/multilingual-e5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()
```

```python
# Функция для получения эмбеддинга документа
def embed_document(text: str, max_length: int = 512) -> np.ndarray:
    input_text = f"passage: {text}"
    inputs = tokenizer(
        input_text,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=max_length
    ).to(device)

    with torch.no_grad():
        outputs = model(**inputs)
        mask = inputs['attention_mask']
        embeddings = outputs.last_hidden_state
        masked = embeddings * mask.unsqueeze(-1)
        summed = masked.sum(dim=1)
        lengths = mask.sum(dim=1, keepdim=True)
        mean_pooled = summed / lengths.clamp(min=1e-9)
        normalized = F.normalize(mean_pooled, p=2, dim=1)
    return normalized.cpu().numpy()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def embed_query(text: str, max_length: int = 512) -> np.ndarray:
    input_text = f"query: {text}"
    inputs = tokenizer(
        input_text,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=max_length
    ).to(device)

    with torch.no_grad():
        outputs = model(**inputs)
        mask = inputs['attention_mask']
        embeddings = outputs.last_hidden_state
        masked = embeddings * mask.unsqueeze(-1)
        summed = masked.sum(dim=1)
        lengths = mask.sum(dim=1, keepdim=True)
        mean_pooled = summed / lengths.clamp(min=1e-9)
        normalized = F.normalize(mean_pooled, p=2, dim=1)
    return normalized.cpu().numpy()
```

```python
def dcg_at_k(relevances, k=5):
    relevances = np.array(relevances[:k])
    if len(relevances) == 0:
        return 0.0
    discounts = np.log2(np.arange(2, relevances.size + 1))
    return relevances[0] + np.sum(relevances[1:] / discounts)
```

```python
def evaluate_e5_model(doc_embeddings, doc_urls, questions_df, k=5):
    doc_urls_clean = [str(url).strip() for url in doc_urls]
    questions_df = questions_df.copy()
    questions_df["correct_url"] = questions_df["correct_url"].astype(str).str.strip()

    recalls_at_1 = []
    recalls_at_5 = []
    dcg_at_5_list = []

    for _, item in tqdm(questions_df.iterrows(), total=len(questions_df)):
        q_emb = embed_query(item["question"])
        scores = cosine_similarity(q_emb, doc_embeddings)[0]

        top_k_indices = np.argsort(-scores)[:k]
        retrieved_urls = [doc_urls_clean[i] for i in top_k_indices]
        correct_url = item["correct_url"]

        # Recall@1
        recall_at_1 = float(retrieved_urls[0] == correct_url)
        recalls_at_1.append(recall_at_1)

        # Recall@5
        recall_at_5 = float(correct_url in retrieved_urls)
        recalls_at_5.append(recall_at_5)

        # DCG@5
        relevances = [1.0 if url == correct_url else 0.0 for url in retrieved_urls]
        dcg = dcg_at_k(relevances, k=k)
        dcg_at_5_list.append(dcg)

    return {
        'recall@1': np.mean(recalls_at_1),
        'recall@5': np.mean(recalls_at_5),
        'dcg@5': np.mean(dcg_at_5_list),
        'recall@1_std': np.std(recalls_at_1),
        'recall@5_std': np.std(recalls_at_5)
    }
```

```python
# --- Оценка на подвыборке ---
qa_subset = questions_and_answers.iloc[:]

metrics_e5 = evaluate_e5_model(doc_embeddings, doc_urls, qa_subset)

print("=== Результаты multilingual-e5-base ===")
print(f"Recall@1: {metrics_e5['recall@1']:.4f}")
print(f"Recall@5: {metrics_e5['recall@5']:.4f}")
print(f"DCG@5:    {metrics_e5['dcg@5']:.4f}")

print("\n=== Сравнение с TF-IDF бейзлайном (title_text) ===")
print("TF-IDF:   Recall@1 = 0.8887, Recall@5 = 0.9624, DCG@5 = 0.9554")
```

```python
import pandas as pd
import numpy as np
import json
import re
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from rank_bm25 import BM25Okapi
from tqdm.notebook import tqdm
```

```python
# Загружаем корпус документов (как в предыдущих заданиях)
with open("swimmers_dataset.json", "r", encoding="utf-8") as f:
    raw_docs = json.load(f)

docs_list = []
for item in raw_docs:
    if item["data"]["sections"]:
        docs_list.append({
            "url": item["url"],
            "title": item["data"]["title"],
            "text": item["data"]["sections"][0]["content"]
        })

docs_df = pd.DataFrame(docs_list)
docs_df["title_text"] = docs_df["title"] + ". " + docs_df["text"]

url_to_doc = dict(zip(docs_df["url"], docs_df["title_text"]))
all_documents = docs_df["title_text"].tolist()
doc_urls = docs_df["url"].tolist()

# Загружаем вопросы
questions_and_answers = pd.read_csv("swimmer_questions_v2.tsv", sep="\t")

# Обеспечим совпадение типов
url_to_doc = {str(k): v for k, v in url_to_doc.items()}
doc_urls = [str(u) for u in doc_urls]
questions_and_answers["correct_url"] = questions_and_answers["correct_url"].astype(str)
```

```python
# 2. Инициализация глобальных признаков

print("Инициализация TF-IDF и BM25...")
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(all_documents)

tokenized_corpus = [doc.split() for doc in all_documents]
bm25 = BM25Okapi(tokenized_corpus)

# Индекс документа для быстрого поиска
doc_to_idx = {doc: i for i, doc in enumerate(all_documents)}

# Загрузка retrieval-модели (E5)
print("Загрузка E5 модели...")
retrieval_model = SentenceTransformer('intfloat/multilingual-e5-base')
```

```python
# 3. Генерация синтетической разметки для reranking
print("Генерация синтетической разметки для reranking...")

rerank_samples = []
for _, row in questions_and_answers.iterrows():
    q = row["question"]
    correct_url = row["correct_url"]
    if correct_url not in url_to_doc:
        continue
    correct_doc = url_to_doc[correct_url]

    # Получаем кандидатов от TF-IDF (бейзлайн)
    q_tfidf = tfidf_vectorizer.transform([q])
    scores = np.array(q_tfidf @ tfidf_matrix.T).flatten()
    top20_idx = np.argsort(-scores)[:20]
    top20_docs = [all_documents[i] for i in top20_idx]

    # Убедимся, что правильный документ в топе (добавим, если нет)
    if correct_doc not in top20_docs:
        top20_docs = [correct_doc] + top20_docs[:19]

    # Создаём 20 пар: (вопрос, документ, label)
    for doc in top20_docs:
        label = 1.0 if doc == correct_doc else 0.0
        rerank_samples.append({
            "question": q,
            "document": doc,
            "label": label
        })

print(f"Создано {len(rerank_samples)} примеров для reranking.")
```

```python
# 4. Вычисление признаков
def compute_features(question: str, document: str) -> np.ndarray:
    # 1. Cosine (E5)
    emb_q = retrieval_model.encode([question], normalize_embeddings=True)
    emb_d = retrieval_model.encode([document], normalize_embeddings=True)
    cosine_sim = float(np.dot(emb_q[0], emb_d[0]))

    # 2. Имя пловца
    name_match = 0.0
    for doc_text in all_documents:
        name = doc_text.split('.')[0] if '.' in doc_text else doc_text[:30]
        if name and name.lower() in question.lower() and name.lower() in document.lower():
            name_match = 1.0
            break

    # 3. Совпадение чисел
    q_nums = set(re.findall(r'\d+', question))
    d_nums = set(re.findall(r'\d+', document))
    num_match = 1.0 if (q_nums & d_nums) else 0.0

    # 4. TF-IDF similarity
    q_vec = tfidf_vectorizer.transform([question])
    d_vec = tfidf_vectorizer.transform([document])
    tfidf_sim = float((q_vec @ d_vec.T).toarray()[0][0])

    # 5. BM25 score
    doc_idx = doc_to_idx.get(document, 0)
    bm25_score = bm25.get_scores(question.split())[doc_idx]
    bm25_norm = min(bm25_score / 10.0, 1.0)

    # 6. Категориальное совпадение
    style_kw = {"брасс", "баттерфляй", "вольный", "спине", "комплекс"}
    country_kw = {"США", "Россия", "Австралия", "Великобритания", "Китай", "Япония", "Германия", "Франция"}
    q_lower, d_lower = question.lower(), document.lower()
    cat_match = 1.0 if (
        (any(kw in q_lower for kw in style_kw) and any(kw in d_lower for kw in style_kw)) or
        (any(kw in q_lower for kw in country_kw) and any(kw in d_lower for kw in country_kw))
    ) else 0.0

    return np.array([
        cosine_sim,
        name_match,
        num_match,
        tfidf_sim,
        bm25_norm,
        cat_match
    ], dtype=np.float32)
```

```python
# 7. Оценка
def evaluate_reranker_on_subset(reranker, qa_subset):
    recalls_at_1 = []
    recalls_at_5 = []
    dcg_at_5_list = []

    for _, row in tqdm(qa_subset.iterrows(), total=len(qa_subset)):
        q = row["question"]
        correct_url = str(row["correct_url"])
        if correct_url not in url_to_doc:
            continue

        # Получаем кандидатов от TF-IDF
        q_tfidf = tfidf_vectorizer.transform([q])
        scores = np.array(q_tfidf @ tfidf_matrix.T).flatten()
        top20_idx = np.argsort(-scores)[:20]
        candidates = [(all_documents[i], doc_urls[i]) for i in top20_idx]

        # Убедимся, что правильный документ в топе
        correct_doc = url_to_doc[correct_url]
        if correct_doc not in [d for d, _ in candidates]:
            candidates = [(correct_doc, correct_url)] + candidates[:19]

        # Извлекаем признаки и предсказываем скоры
        features_list = []
        for doc, _ in candidates:
            feat = compute_features(q, doc)
            features_list.append(feat)
        features_array = np.vstack(features_list)
        scores_rerank = reranker.predict_proba(features_array)[:, 1]

        # Сортируем
        reranked = sorted(zip(candidates, scores_rerank), key=lambda x: x[1], reverse=True)
        retrieved_urls = [url for (_, url), _ in reranked[:5]]

        # Метрики
        recalls_at_1.append(1.0 if retrieved_urls[0] == correct_url else 0.0)
        recalls_at_5.append(1.0 if correct_url in retrieved_urls else 0.0)

        relevances = [1.0 if url == correct_url else 0.0 for url in retrieved_urls]
        dcg = relevances[0] + sum(rel / np.log2(i + 1) for i, rel in enumerate(relevances[1:], start=2))
        dcg_at_5_list.append(dcg)

    return {
        "recall@1": np.mean(recalls_at_1),
        "recall@5": np.mean(recalls_at_5),
        "dcg@5": np.mean(dcg_at_5_list)
    }

qa_subset = questions_and_answers.iloc[:]
metrics_rerank = evaluate_reranker_on_subset(reranker, qa_subset)

print("\n=== Результаты переранжирования (Logistic + 6 факторов) ===")
print(f"Recall@1: {metrics_rerank['recall@1']:.4f}")
print(f"Recall@5: {metrics_rerank['recall@5']:.4f}")
print(f"DCG@5:    {metrics_rerank['dcg@5']:.4f}")

print("\n=== TF-IDF бейзлайн (title_text) ===")
print("Recall@1 = 0.8887, Recall@5 = 0.9624, DCG@5 = 0.9554")
```

```python
def create_llm_answer(message, model, tokenizer, enable_thinking=False):
    messages = [
        {"role": "user", "content": message}
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=enable_thinking
    )

    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

    # conduct text completion
    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=4000
    )
    output_ids = generated_ids[0][len(model_inputs.input_ids[0]):].tolist()

    # parsing thinking content
    try:
        # rindex finding 151668 (</think>)
        index = len(output_ids) - output_ids[::-1].index(151668)
    except ValueError:
        index = 0

    thinking_content = tokenizer.decode(output_ids[:index], skip_special_tokens=True).strip("\n")
    content = tokenizer.decode(output_ids[index:], skip_special_tokens=True).strip("\n")

    return content
```

```python
from tqdm.notebook import tqdm
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# Инициализация TF-IDF для получения кандидатов (этап 1: retrieval)
all_texts = data["title_text"].tolist()
doc_urls_list = data["url"].tolist()

# Если TF-IDF ещё не обучен — обучите его (должен совпадать с бейзлайном)
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(all_texts)

# Маппинг для быстрого доступа к тексту по URL
url_to_text = dict(zip(data["url"], data["text"]))

# Функция переранжирования
def rerank_with_model(question: str, candidate_urls: list, reranker) -> list:
    """Возвращает отранжированный список URL (от наиболее релевантного к наименее)"""
    candidate_docs = [data.loc[data["url"] == url, "title_text"].iloc[0] for url in candidate_urls]
    features = np.vstack([compute_features(question, doc) for doc in candidate_docs])
    scores = reranker.predict_proba(features)[:, 1]  # вероятность релевантности
    # Сортируем URL по убыванию скоров
    reranked = [url for _, url in sorted(zip(scores, candidate_urls), reverse=True)]
    return reranked

llm_answers = []
MAX_QUESTIONS_TO_PROCESS = 638
start_from = 0

for idx, item in tqdm(
    questions_and_answers.iloc[start_from:start_from + MAX_QUESTIONS_TO_PROCESS].iterrows(),
    total=MAX_QUESTIONS_TO_PROCESS
):
    question = item["question"]
    correct_url = item["correct_url"]

    # 1. Получаем кандидатов от TF-IDF (бейзлайн retrieval)
    q_tfidf = tfidf_vectorizer.transform([question])
    tfidf_scores = np.array(q_tfidf @ tfidf_matrix.T).flatten()
    top20_indices = np.argsort(-tfidf_scores)[:20]
    candidate_urls = [doc_urls_list[i] for i in top20_indices]

    # Гарантируем наличие правильного документа (для честной оценки)
    if correct_url not in candidate_urls:
        candidate_urls[-1] = correct_url  # заменяем последний

    # 2. ПЕРЕРАНЖИРУЕМ кандидатов с помощью reranker
    reranked_urls = rerank_with_model(question, candidate_urls, reranker)
    top5_urls = reranked_urls[:5]

    # 3. Метрики
    recall_at_1 = 1 if top5_urls[0] == correct_url else 0
    recall_at_5 = 1 if correct_url in top5_urls else 0

    # 4. Генерация ответов

    # Без RAG
    llm_answer = create_llm_answer(question, model, tokenizer)

    # RAG: 1 документ (после reranking)
    top1_doc_text = url_to_text[top5_urls[0]]
    message_rag1 = (
        f"Ответь на вопрос, используя только информацию ниже.\n\n"
        f"Контекст:\n{top1_doc_text}\n\n"
        f"Вопрос: {question}"
    )
    llm_answer_rag1 = create_llm_answer(message_rag1, model, tokenizer)

    # RAG: 5 документов (после reranking)
    rag_docs = []
    for i, url in enumerate(top5_urls, 1):
        doc_text = url_to_text[url]
        rag_docs.append(f"Документ {i}:\n{doc_text}")
    message_rag5 = (
        f"Ответь на вопрос, используя информацию из следующих документов.\n\n"
        + "\n\n".join(rag_docs) + "\n\n"
        + f"Вопрос: {question}"
    )
    llm_answer_rag5 = create_llm_answer(message_rag5, model, tokenizer)

    # 5. Сохранение
    llm_answers.append({
        "question": question,
        "correct_answer": item["correct_answer"],
        "llm_answer": llm_answer,
        "llm_answer_rag1": llm_answer_rag1,
        "llm_answer_rag5": llm_answer_rag5,
        "is_top1_doc_rel": bool(recall_at_1),
        "has_top5_doc_rel": bool(recall_at_5),
        "correct_url": correct_url,
        "retrieved_urls_top5": "; ".join(top5_urls)
    })

# Сохранение
df_to_annotate = pd.DataFrame(llm_answers)
df_to_annotate.to_excel("llm_rag_annotation_task_reranker.xlsx", index=False)
print("Файл сохранён: llm_rag_annotation_task_reranker.xlsx")
```

```python
import re

def get_judge_score_final(question: str, reference_answer: str, model_answer: str, model, tokenizer) -> str:

    prompt = f"""Ты — строгий эксперт-оценщик. Твоя задача — сравнить два ответа и вывести ТОЛЬКО ОДНУ ЦИФРУ без каких-либо объяснений.


Пример:
ВОПРОС: Сколько золотых медалей у Майкла Фелпса?
ЭТАЛОННЫЙ ОТВЕТ: 23
ОТВЕТ МОДЕЛИ: У Майкла Фелпса 23 золотые медали.
Цифра: 1

Пример:
ВОПРОС: В каком году родился Иан Торп?
ЭТАЛОННЫЙ ОТВЕТ: 1982
ОТВЕТ МОДЕЛИ: Он родился в конце 1983.
Цифра: 0

Пример:
ВОПРОС: Кто стала чемпионкой олимпиады 2012 в Лондоне по плаванию вольным стилемм 50 м среди женщин?
ЭТАЛОННЫЙ ОТВЕТ: Раноми Кромовидьойо
ОТВЕТ МОДЕЛИ: Это был Майкл Фелпс.
Цифра: 0


ВОПРОС: {question}

ЭТАЛОННЫЙ ОТВЕТ: {reference_answer}

ОТВЕТ МОДЕЛИ: {model_answer}

Если ответ модели СУЩЕСТВЕННО соответствует эталонному ответу по смыслу и содержит правильную информацию, выведи: 1
Если ответ модели НЕ соответствует эталону, содержит ошибки или неточности, выведи: 0

Цифра:"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=5,
        do_sample=False,
        temperature=0.0,
        pad_token_id=tokenizer.eos_token_id
    )

    generated_tokens = outputs[0][inputs.input_ids.shape[1]:]
    generated_text = tokenizer.decode(generated_tokens, skip_special_tokens=True).strip()


    for char in generated_text:
        if char == '0':
            return '0'
        elif char == '1':
            return '1'
        elif char.isspace():
            continue
        else:
            break

    match = re.search(r'^\s*([01])', generated_text)
    if match:
        return match.group(1)
    return 'ERROR'
```

```python
import pandas as pd
import re

# Загрузка размеченных данных
df = pd.read_excel("llm_rag_annotation_task_reranker.xlsx")

def safe_judge(*args, **kwargs):
    """Обёртка для обработки ошибок и некорректных выходов"""
    result = get_judge_score_final(*args, **kwargs)
    if result in ('0', '1'):
        return int(result)
    else:
        print(f" Судья вернул некорректное значение: '{result}' → заменено на 0")
        return 0  # или np.nan, если предпочитаете
```

```python
# Применяем LLM-as-a-judge ко всем трём типам ответов
print("Оценка ответов без RAG...")
df["llm_judge_no_rag"] = df.apply(
    lambda row: safe_judge(row["question"], row["correct_answer"], row["llm_answer"], model, tokenizer),
    axis=1
)

print("Оценка ответов с RAG-1...")
df["llm_judge_rag1"] = df.apply(
    lambda row: safe_judge(row["question"], row["correct_answer"], row["llm_answer_rag1"], model, tokenizer),
    axis=1
)

print("Оценка ответов с RAG-5...")
df["llm_judge_rag5"] = df.apply(
    lambda row: safe_judge(row["question"], row["correct_answer"], row["llm_answer_rag5"], model, tokenizer),
    axis=1
)

# Сохраняем результат
df.to_excel("llm_rag_with_judge_scores_task_10.xlsx", index=False)
print("Файл сохранён: llm_rag_with_judge_scores_task_10.xlsx")
```
